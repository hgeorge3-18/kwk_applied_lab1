### HfN film - GIXRD data 

```python
import numpy as np
def read_csv(fname):
    headers = np.genfromtxt(fname, delimiter = ",", max_rows = 1, dtype = str) 
    data = np.genfromtxt(fname, delimiter = ",", skip_header = 1, dtype = float)
    return headers.tolist(), data 
```

```python
headers, data = read_csv('data/HfN_170nm_GIXRD.csv')

print(headers) 
print(data)
```

```python
import matplotlib.pyplot as mpl
import numpy as np 

headers, data = read_csv('data/HfN_170nm_GIXRD.csv')

angle_20 = data [:,0]
intensity = data[:,1]

mpl.plot( angle_20, intensity)

mpl.title("GIXRD of HfN Film of 170 nm")
mpl.xlabel('2 theta (degrees)')
mpl.ylabel('Intensity (cps)')

# phases 
mpl.text(32, 1000, "(111)", fontsize = 10, rotation = 90)
mpl.text(37, 900, "(200)", fontsize = 10, rotation = 90)
mpl.text(54.75, 350, "(220)", fontsize = 10, rotation = 90)
mpl.text(65.5, 225, "(311)", fontsize = 10, rotation = 90)

mpl.show()
```
